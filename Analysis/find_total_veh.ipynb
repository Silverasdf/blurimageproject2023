```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('MasterSheet_2023_updated.csv')
df2 = pd.read_csv('MasterSheet_CSV_updated.csv')
df2
```

```python
#Grab vehicle ids - 2023
ids = {}
for i, row in df.iterrows():
    vid = row["Filename"].split('_')[1]
    if vid not in ids:
        ids[vid] = 1
    else:
        ids[vid] += 1
print(ids)
```

```python
#Grab vehicle ids - CSV
ids2 = {}
for i, row in df2.iterrows():
    try:
        vid = row["Filename"].split('_')[1][8:]
        #print(vid)
    except:
        continue
    if vid not in ids2:
        ids2[vid] = 1
    else:
        ids2[vid] += 1
    if row['MatchFilename'] != '0' and row['MatchFilename'] != '1' and row['MatchFilename'] != 0 and row['MatchFilename'] != 1:
        ids2[vid] += 1
print(ids2)
```

```python
#Find total number of ids and median value of dict
values = np.array([ids[key] for key in ids])
values2 = np.array([ids2[key] for key in ids2])
```

```python
print(len(ids.keys()))
print(len(ids2.keys()))
print(np.median(values))
print(np.median(values2))
```
